```python
# Definir pesos para cada indicador
pesos = {
    'Media_Movel_Lucro': 3.0,
    'Taxa_Crescimento_Lucro': 2.5,
    'Divida_Sobre_Patrimonio': 1.5,
    'Retorno_Sobre_Ativos': 2.0,
    'Margem_de_Lucro': 2.0,
}

# Calcular a pontuação para cada empresa
def calcular_pontuacao(row):
    pontuacao = 0
    for chave, peso in pesos.items():
        if chave in row and pd.notnull(row[chave]):
            pontuacao += row[chave] * peso
    return pontuacao

dados_anuais['Pontuacao'] = dados_anuais.apply(calcular_pontuacao, axis=1)
```

```python
# Filtrar dados para o ano mais recente para análise
ano_recente = dados_anuais['Ano'].max()
dados_atual = dados_anuais[dados_anuais['Ano'] == ano_recente]

# Ordenar empresas com base na pontuação
dados_atual = dados_atual.sort_values(by='Pontuacao', ascending=False).reset_index(drop=True)

# Definir limiares para classificação
def classificar_acao(pontuacao):
    if pontuacao >= 8:
        return 'Compra Forte'
    elif pontuacao >= 6:
        return 'Compra'
    elif pontuacao >= 4:
        return 'Manter'
    elif pontuacao >= 2:
        return 'Venda'
    else:
        return 'Venda Forte'

dados_atual['Recomendacao'] = dados_atual['Pontuacao'].apply(classificar_acao)

# Selecionar as colunas desejadas, incluindo 'cod_negociacao'
principais_recomendacoes = dados_atual[['CNPJ', 'DENOM_CIA', 'cod_negociacao', 'Pontuacao', 'Recomendacao']]

# Verificar a posição da WEGE3 no ranking geral
posicao_wege3 = principais_recomendacoes[principais_recomendacoes['cod_negociacao'] == 'WEGE3'].index
if len(posicao_wege3) > 0:
    print(f"A WEGE3 está na posição: {posicao_wege3[0] + 1} no ranking geral.")
else:
    print("A WEGE3 não foi encontrada no ranking geral.")
```
