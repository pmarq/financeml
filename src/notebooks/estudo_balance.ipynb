```python

##NOVAS CONTAS##

# Supondo que df_cleaned já esteja definido e contenha as colunas mencionadas

# Supondo que df_cleaned já esteja definido e contenha as colunas mencionadas

# 1. Calcular Passivo Total (se ainda não estiver calculado)
df_cleaned['Passivo Total'] = df_cleaned['2.01 - Passivo Circulante'].fillna(0) + df_cleaned['2.02 - Passivo Não Circulante'].fillna(0)

# 2. Calcular ROE (Return on Equity)
df_cleaned['ROE (%)'] = (df_cleaned['Lucro/Prejuízo Consolidado'] / df_cleaned['Patrimonio_Liquido_Unificado']) * 100

# 3. Calcular ROA (Return on Assets)
df_cleaned['ROA (%)'] = (df_cleaned['Lucro/Prejuízo Consolidado'] / df_cleaned['1 - Ativo Total']) * 100

# 4. Calcular Liquidez Corrente
df_cleaned['Liquidez Corrente'] = df_cleaned['1.01 - Ativo Circulante'] / df_cleaned['2.01 - Passivo Circulante']

# 5. Calcular Grau de Endividamento (Debt to Equity Ratio)
df_cleaned['Endividamento'] = df_cleaned['Passivo Total'] / df_cleaned['Patrimonio_Liquido_Unificado']

# 6. Calcular Cobertura de Passivos
df_cleaned['Cobertura de Passivos'] = df_cleaned['1.01 - Ativo Circulante'] / df_cleaned['Passivo Total']

# 7. Calcular ROC (Return on Capital)
df_cleaned['ROC (%)'] = (df_cleaned['Lucro/Prejuízo Consolidado'] / 
                          (df_cleaned['1 - Ativo Total'] - df_cleaned['Passivo Total'])) * 100

# 8. Calcular Equity Ratio
df_cleaned['Equity_Ratio'] = df_cleaned['Patrimonio_Liquido_Unificado'] / df_cleaned['1 - Ativo Total']

# 9. Calcular Debt to Asset Ratio
df_cleaned['Debt_to_Asset_Ratio'] = df_cleaned['Passivo Total'] / df_cleaned['1 - Ativo Total']

# 10. Calcular Capital de Giro (Working Capital)
df_cleaned['Working_Capital'] = df_cleaned['1.01 - Ativo Circulante'] - df_cleaned['2.01 - Passivo Circulante']

# 11. Calcular Solvency Ratio
df_cleaned['Solvency_Ratio'] = df_cleaned['1 - Ativo Total'] / df_cleaned['Passivo Total']

# 12. Calcular Índice de Endividamento de Longo Prazo (Long-term Debt Ratio)
df_cleaned['Long_Term_Debt_Ratio'] = df_cleaned['2.02 - Passivo Não Circulante'] / df_cleaned['1 - Ativo Total']

# 13. Calcular Índice de Endividamento de Curto Prazo (Short-term Debt Ratio)
df_cleaned['Short_Term_Debt_Ratio'] = df_cleaned['2.01 - Passivo Circulante'] / df_cleaned['1 - Ativo Total']

# 14. Calcular Índice de Eficiência Operacional (Operating Efficiency Ratio)
df_cleaned['Operating_Efficiency_Ratio'] = df_cleaned['Resultado Operacional Consolidado'] / df_cleaned['1 - Ativo Total']

# 15. Calcular Operating Profit Margin
df_cleaned['Operating_Profit_Margin (%)'] = (df_cleaned['Resultado Operacional Consolidado'] / df_cleaned['1 - Ativo Total']) * 100

# Indicadores Adicionais

# 16. Calcular Equity Multiplier
df_cleaned['Equity_Multiplier'] = df_cleaned['1 - Ativo Total'] / df_cleaned['Patrimonio_Liquido_Unificado']

# 17. Calcular Capital Employed
df_cleaned['Capital_Employed'] = df_cleaned['1 - Ativo Total'] - df_cleaned['2.01 - Passivo Circulante']

# 18. Calcular Debt to Capital Ratio
df_cleaned['Debt_to_Capital_Ratio'] = df_cleaned['Passivo Total'] / (df_cleaned['Passivo Total'] + df_cleaned['Patrimonio_Liquido_Unificado'])

# 19. Calcular Capital to Debt Ratio
df_cleaned['Capital_to_Debt_Ratio'] = df_cleaned['Patrimonio_Liquido_Unificado'] / df_cleaned['Passivo Total']

# 20. Calcular Profitability Index
df_cleaned['Profitability_Index'] = df_cleaned['Lucro/Prejuízo Consolidado'] / df_cleaned['1 - Ativo Total']

# 21. Calcular Financial Leverage Ratio
df_cleaned['Financial_Leverage_Ratio'] = df_cleaned['1 - Ativo Total'] / df_cleaned['Patrimonio_Liquido_Unificado']

# Exibir as primeiras linhas da tabela consolidada com os novos indicadores
df_cleaned.head()
```

```python

# Convert date columns to datetime
df_balance_dre_ind['DT_FIM_EXERC'] = pd.to_datetime(df_balance_dre_ind['DT_FIM_EXERC'])
cotacao_hist['data_pregao'] = pd.to_datetime(cotacao_hist['data_pregao'])

# Function to find the nearest available date
def encontrar_data_proxima(data, datas_disponiveis):
    datas_diferenca = datas_disponiveis - data
    datas_diferenca_pos = datas_diferenca[datas_diferenca >= pd.Timedelta(0)]
    if not datas_diferenca_pos.empty:
        return datas_disponiveis[datas_diferenca_pos.idxmin()]
    return datas_disponiveis[datas_diferenca.abs().idxmin()]

# Transform the unique date series into a list for efficient comparison
datas_unicas = pd.Series(cotacao_hist['data_pregao'].unique())

# Apply the function to find the nearest available date for each record
df_balance_dre_ind['data_pregao'] = df_balance_dre_ind['DT_FIM_EXERC'].apply(lambda x: encontrar_data_proxima(x, datas_unicas))

# Merge the two datasets
df_merged = pd.merge(
    df_balance_dre_ind, 
    cotacao_hist[['CNPJ', 'cod_negociacao', 'data_pregao', 'preco_ultimo_negocio']],
    left_on=['CNPJ', 'data_pregao'], 
    right_on=['CNPJ', 'data_pregao'], 
    how='left'
)

# Sort by CNPJ and DT_FIM_EXERC to prepare for percentage change calculation
df_merged_sorted = df_merged.sort_values(by=['CNPJ', 'DT_FIM_EXERC'])

# Calculate the percentage change in stock prices for each company
df_merged_sorted['variacao_percentual'] = df_merged_sorted.groupby('CNPJ')['preco_ultimo_negocio'].pct_change() * 100

# Remove rows where percentage variation is NaN (resulting from pct_change)
df_merged_sorted.dropna(subset=['variacao_percentual'], inplace=True)

# Optional: Display or save the final dataframe
df_merged_sorted

```

```python
""" A seguir, detalhamos cada um dos indicadores calculados, incluindo os adicionais sugeridos:

ROE (%) - Return on Equity (Retorno sobre o Patrimônio Líquido)
Fórmula: (Lucro/Prejuízo Consolidado / Patrimonio_Liquido_Unificado) * 100
Descrição: Mede a rentabilidade do patrimônio líquido, indicando quanto lucro a empresa gera com os recursos dos acionistas.
Importância: Avalia a eficiência da gestão em utilizar o capital próprio para gerar lucros.

ROA (%) - Return on Assets (Retorno sobre os Ativos)
Fórmula: (Lucro/Prejuízo Consolidado / 1 - Ativo Total) * 100
Descrição: Avalia a eficiência da empresa em utilizar seus ativos totais para gerar lucro.
Importância: Indica quão bem a empresa está convertendo ativos em lucros.

Liquidez Corrente - Current Ratio
Fórmula: 1.01 - Ativo Circulante / 2.01 - Passivo Circulante
Descrição: Avalia a capacidade da empresa de cumprir suas obrigações de curto prazo com seus ativos circulantes.
Importância: Indica a saúde de curto prazo da empresa e sua capacidade de pagar dívidas imediatas.

Endividamento - Debt to Equity Ratio (Dívida sobre Patrimônio Líquido)
Fórmula: Passivo Total / Patrimonio_Liquido_Unificado
Descrição: Mede a proporção de recursos de terceiros (dívidas) em relação aos recursos próprios dos acionistas.
Importância: Avalia a estrutura de capital da empresa e seu nível de alavancagem financeira.

Cobertura de Passivos - Coverage of Liabilities
Fórmula: 1.01 - Ativo Circulante / Passivo Total
Descrição: Avalia a capacidade da empresa de cobrir todas as suas obrigações financeiras com seus ativos circulantes.
Importância: Indica a segurança financeira da empresa em relação às suas dívidas totais.

ROC (%) - Return on Capital (Retorno sobre o Capital)
Fórmula: (Lucro/Prejuízo Consolidado / (1 - Ativo Total - Passivo Total)) * 100
Descrição: Mede a eficiência da empresa em gerar lucro a partir do capital investido, considerando o patrimônio líquido.
Importância: Avalia a rentabilidade do capital total empregado na empresa.

Equity Ratio - Índice de Participação do Patrimônio Líquido
Fórmula: Patrimonio_Liquido_Unificado / 1 - Ativo Total
Descrição: Indica a proporção dos ativos que são financiados pelo patrimônio líquido dos acionistas.
Importância: Avalia a solidez financeira e a independência da empresa em relação a financiamentos externos.

Debt to Asset Ratio - Índice de Endividamento sobre Ativos
Fórmula: Passivo Total / 1 - Ativo Total
Descrição: Mede a proporção dos ativos da empresa que são financiados por dívidas.
Importância: Avalia a alavancagem financeira da empresa e seu risco associado.

Working Capital - Capital de Giro
Fórmula: 1.01 - Ativo Circulante - 2.01 - Passivo Circulante
Descrição: Mede a capacidade da empresa de financiar suas operações de curto prazo.
Importância: Indica se a empresa possui recursos suficientes para cobrir suas obrigações de curto prazo.

Solvency Ratio - Índice de Solvência
Fórmula: 1 - Ativo Total / Passivo Total
Descrição: Avalia a capacidade da empresa de cumprir todas as suas obrigações financeiras com seus ativos totais.
Importância: Indica a estabilidade financeira e a capacidade de sobrevivência a longo prazo da empresa.

Long-term Debt Ratio - Índice de Endividamento de Longo Prazo
Fórmula: 2.02 - Passivo Não Circulante / 1 - Ativo Total
Descrição: Mostra a proporção dos ativos financiados por dívidas de longo prazo.
Importância: Ajuda a entender a estrutura de financiamento de longo prazo da empresa.

Short-term Debt Ratio - Índice de Endividamento de Curto Prazo
Fórmula: 2.01 - Passivo Circulante / 1 - Ativo Total
Descrição: Mede a proporção dos ativos financiados por dívidas de curto prazo.
Importância: Indica a pressão sobre a liquidez da empresa devido a dívidas de curto prazo.

Operating Efficiency Ratio - Índice de Eficiência Operacional
Fórmula: Resultado Operacional Consolidado / 1 - Ativo Total
Descrição: Avalia quão eficientemente a empresa está utilizando seus ativos para gerar resultado operacional.
Importância: Indica a eficácia das operações da empresa na geração de lucro a partir dos ativos.

Operating Profit Margin (%) - Margem de Lucro Operacional
Fórmula: (Resultado Operacional Consolidado / 1 - Ativo Total) * 100
Descrição: Mede a margem de lucro operacional em relação aos ativos totais.
Importância: Indica a eficiência operacional e a capacidade da empresa de gerar lucro a partir de suas operações principais.

Equity Multiplier - Multiplicador de Patrimônio
Fórmula: 1 - Ativo Total / Patrimonio_Liquido_Unificado
Descrição: Mede a alavancagem financeira da empresa, indicando quanto dos ativos são financiados por recursos próprios.
Importância: Avalia o grau de dependência da empresa em relação a financiamentos externos.

Capital Employed - Capital Empregado
Fórmula: 1 - Ativo Total - 2.01 - Passivo Circulante
Descrição: Representa os recursos utilizados para financiar as operações da empresa, excluindo as obrigações de curto prazo.
Importância: Indica quanto do capital total está sendo utilizado para operações de longo prazo.

Debt to Capital Ratio - Índice de Dívida sobre Capital
Fórmula: Passivo Total / (Passivo Total + Patrimonio_Liquido_Unificado)
Descrição: Mede a proporção de dívida em relação ao capital total da empresa.
Importância: Avalia a composição do financiamento da empresa e seu nível de alavancagem financeira.

Capital to Debt Ratio - Índice de Capital sobre Dívida
Fórmula: Patrimonio_Liquido_Unificado / Passivo Total
Descrição: Indica a proporção de capital próprio em relação às dívidas totais da empresa.
Importância: Avalia a capacidade da empresa de suportar suas dívidas com seus próprios recursos.

Profitability Index - Índice de Rentabilidade
Fórmula: Lucro/Prejuízo Consolidado / 1 - Ativo Total
Descrição: Mede a capacidade da empresa de gerar lucro a partir de seus ativos totais.
Importância: Avalia a eficiência na geração de lucro com os ativos disponíveis.

Financial Leverage Ratio - Índice de Alavancagem Financeira
Fórmula: 1 - Ativo Total / Patrimonio_Liquido_Unificado
Descrição: Mede o grau de alavancagem financeira da empresa, indicando quanto dos ativos são financiados por recursos próprios.
Importância: Avalia o nível de risco financeiro associado à estrutura de capital da empresa. """
```
