```python
df_dradata_tri = df_dradata_tri.drop(columns=['VERSAO', 'MOEDA', 'ORDEM_EXERC', 'ST_CONTA_FIXA','ESCALA_MOEDA','DT_REFER','DT_INI_EXERC'])
df_dradata_tri.info()
```

```python
# Pivotando o DataFrame para transformar valores de DS_CONTA em colunas
df_pivot_dra_tri = df_dradata_tri.pivot_table(index=['CNPJ_CIA', 'DENOM_CIA', 'CD_CVM', 'GRUPO_DFP', 'DT_FIM_EXERC'], 
                          columns='DS_CONTA', 
                          values='VL_CONTA', 
                          aggfunc='sum')  # Escolha 'sum', 'mean' ou outra função conforme necessário

df_pivot_dra_tri
```

```python
# Reseta o índice para deixar as colunas de índice como colunas normais, se necessário
df_pivot_dra_tri.reset_index(inplace=True)

df_pivot_dra_tri 
```

```python
df_pivot_dra_tri.isnull().sum()
```

```python
# Calcular o limite para manter as colunas (total de linhas menos 200)
limite = len(df_pivot_dra_tri) - 200

# Remover as colunas que têm mais de 200 valores nulos
df_pivot_dra_tri = df_pivot_dra_tri.dropna(thresh=limite, axis=1)

# Verificar as colunas restantes
print(df_pivot_dra_tri.info())
```

```python
pd.DataFrame(df_pivot_dra_tri)
```

```python
# Cria uma máscara para encontrar onde não há valores nulos em nenhuma coluna
mask_no_nulls = df_pivot_dra_tri.notnull().all(axis=1)

# Aplica a máscara para obter apenas as linhas sem nenhum valor nulo
empresas_sem_nulos = df_pivot_dra_tri[mask_no_nulls]

# Exibe as empresas sem valores ausentes
print(empresas_sem_nulos[['CNPJ_CIA', 'DENOM_CIA']])
```

```python
# Contando o número de CNPJs únicos no DataFrame
numero_de_cnpjs_unicos = empresas_sem_nulos['CNPJ_CIA'].nunique()

# Exibindo o resultado
print("Número de CNPJs únicos:", numero_de_cnpjs_unicos)
```

```python
empresas_sem_nulos.columns
```

```python
df = pd.DataFrame(empresas_sem_nulos)
```

```python
# Análise de Desempenho Abrangente
df['Relação Lucro/Resultado Abrangente'] = df['Lucro Líquido do Período'] / df['Resultado Abrangente do Período']

# Verificando mudanças ano a ano se dados temporais estão disponíveis
df['Variação Anual do Lucro Líquido'] = df['Lucro Líquido do Período'].pct_change()
df['Variação Anual do Resultado Abrangente'] = df['Resultado Abrangente do Período'].pct_change()

# Exibindo as novas análises
print(df[['Relação Lucro/Resultado Abrangente', 'Variação Anual do Lucro Líquido', 'Variação Anual do Resultado Abrangente']])
```
