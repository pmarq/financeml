```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Especificar o caminho do arquivo CSV
arquivo_csv = r"C:\Users\usuario\Desktop\Programacao\financeml\src\data\dados_tratados\COTAHIST_TODOS_ANOS.csv"

# Ler o arquivo CSV com a codificaÃ§Ã£o apropriada
preco_acoes_todos_os_anos = pd.read_csv(arquivo_csv, encoding='latin1')

# Exibir as primeiras linhas do dataframe lido
pd.DataFrame(preco_acoes_todos_os_anos)
```

```python
# Filtrando apenas as colunas desejadas
colunas_desejadas = ['data_pregao', 'cod_negociacao', 'noma_empresa', 'preco_ultimo_negocio']
preco_acoes = preco_acoes_todos_os_anos[colunas_desejadas]

# Filtrando apenas as empresas com cod_negociacao terminando em '3'
preco_acoes = preco_acoes[preco_acoes['cod_negociacao'].str.endswith('3')]

preco_acoes_final3 = preco_acoes 

# Exibindo o DataFrame filtrado
pd.DataFrame(preco_acoes_final3)
```

```python
# Filtrando apenas os cÃ³digos de negociaÃ§Ã£o que terminam exatamente com '3'
preco_acoes_final3= preco_acoes_final3[preco_acoes_final3['cod_negociacao'].str.match(r'^[A-Z]{4}3$')]

pd.DataFrame(preco_acoes_final3)
```

```python
# Calculando a variaÃ§Ã£o percentual diÃ¡ria
preco_acoes_final3['percentual_variacao'] = preco_acoes_final3.groupby('cod_negociacao')['preco_ultimo_negocio'].pct_change() * 100

# Calculando a mÃ©dia da variaÃ§Ã£o percentual diÃ¡ria para cada empresa
df_variacao_media = preco_acoes_final3.groupby('cod_negociacao')['percentual_variacao'].mean().reset_index()

# Ordenando as empresas pela mÃ©dia da variaÃ§Ã£o percentual em ordem decrescente
df_variacao_media_sorted = df_variacao_media.sort_values(by='percentual_variacao', ascending=False)

pd.DataFrame(df_variacao_media_sorted)
```

```python
top_100_empresas = df_variacao_media_sorted.head(100)
pd.DataFrame(top_100_empresas)
```

```python
# Convertendo a coluna 'data_pregao' para o formato de data
preco_acoes_final3['data_pregao'] = pd.to_datetime(preco_acoes_final3['data_pregao'], format='%Y%m%d')

# Filtrando os dados dos Ãºltimos 5 anos
data_limite = pd.Timestamp.now() - pd.DateOffset(years=5)
df_historico_5_anos = preco_acoes_final3[preco_acoes_final3['data_pregao'] >= data_limite]

# Calculando a variaÃ§Ã£o percentual diÃ¡ria
df_historico_5_anos['percentual_variacao'] = df_historico_5_anos.groupby('cod_negociacao')['preco_ultimo_negocio'].pct_change() * 100

# Calculando a mÃ©dia da variaÃ§Ã£o percentual diÃ¡ria para cada empresa
df_variacao_media = df_historico_5_anos.groupby('cod_negociacao')['percentual_variacao'].mean().reset_index()

# Ordenando as empresas pela mÃ©dia da variaÃ§Ã£o percentual em ordem decrescente
df_variacao_media_sorted = df_variacao_media.sort_values(by='percentual_variacao', ascending=False)

# Selecionando as top 100 empresas (ou menos, se nÃ£o houver 100 empresas no exemplo)
top_50_empresas = df_variacao_media_sorted.head(50)
top_50_empresas_ultimos5anos = top_100_empresas

pd.DataFrame(top_50_empresas_ultimos5anos)
```
